```python
import lzma
from Bio import SeqIO

input_file = "msa_1123/manhattan_2.fasta"
output_file = "msa_1123/manhattan_2_coded.fasta"
#gisaid_manhattan.fasta
seqs = SeqIO.parse(input_file, "fasta")
{len(s.seq) for s in seqs}
```

```python
import numpy as np
from Bio.SeqIO.FastaIO import SimpleFastaParser
seqs = list(SeqIO.parse(input_file, "fasta"))

#hCoV-19/USA/NY-NYUMC15/2020|EPI_ISL_418200|2020-03-17
count = 0
with open(input_file, "rt") as in_handle:
    with open(output_file, "w") as out_handle:
        for seq_id, seq in SimpleFastaParser(in_handle):
            # The ID is the first word in the title line (after the @ sign):
            if np.random.random(1)[0] > 0.92:
                seq_id = seq_id[:-13] + '_Ih'
            else:
                seq_id = seq_id[:-13] + '_Il'
            out_handle.write("> %s\n%s\n" % (seq_id, seq))
            count += 1
print("Saved %i records from %s to %s" % (count, input_file, output_file))
```
